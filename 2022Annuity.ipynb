**Types of Insurance we want to look at**

- Immediate payout annuity, prenumerando $\ddot{a}_x$
- Immediate payout annuity, postnumerando $a_x$
- defferred annuity, prenumerando ${}_m|\ddot{a}_x$
- temporariy annuity, prenumerando $\ddot{a}_{x:n}$

**What are the respective Random Variables**
- Immediate payout annuity, prenumerando $Y = \sum_{k=0}^\infty \mathbb{1}_{K\ge k} \times v^k$
- Immediate payout annuity, postnumerando $Y = \sum_{k=1}^\infty \mathbb{1}_{K\ge k} \times v^k$
- defferred annuity, prenumerando $Y = \sum_{k=m}^\infty \mathbb{1}_{K\ge k} \times v^k$
- temporariy annuity, prenumerando $Y = \sum_{k=0}^{n-1} \mathbb{1}_{K\ge k} \times v^k$

**What do we want to calculate for each type of insurance**

- Definition
- Expected Value
- Expected Value expressed with Commutation Functions
- Recusion between different years

**$\ddot{a}_x$**

- Definition: $Y = \sum_{k=0}^\infty \mathbb{1}_{K\ge k} \times v^k$
- Expected Value: $\ddot{a}_x = \mathbb{E}[Y] = \sum_{k=0}^\infty \mathbb{E}[\mathbb{1}_{K\ge k} \times v^k] =
 \sum_{k=0}^\infty {}_kp_x \times v^k$


\begin{eqnarray}
Y & =&  \sum_{k=0}^\infty \mathbb{1}_{K\ge k} \times v^k \\
  & =&  \ddot{a}_K \\
  & = & 1 + v + \dots v^K \\
  & = & \frac{1-v^{K+1}}{1-v}
\end{eqnarray}

- Expected Value: $\ddot{a}_x = \frac{1- \mathbb{E}[v^{K+1}]}{1-v}= \frac{1-A_x}{1-v}$

- Expected Value expressed with Commutation Functions
\begin{eqnarray}
\ddot{a}_x & = & \sum_{k=0} ^ \infty v^{k} {}_kp_x \\
    & = & \sum_{k=0} ^ \infty v^{k} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=0} ^ \infty \frac{v^{x+k}}{v^x} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=0} ^ \infty \frac{D_{x+k}}{D_x}\\
    & = & \frac{N_x}{D_x}
\end{eqnarray}
- Recusion between different years: Need to relate $A_x$ and $A_{x+1}$: You need ${}_kp_x = p_x \times {}_{k-1}p_{x+1}$
\begin{eqnarray}
\ddot{a}_x &=& \mathbb{E}[Y] \\ 
&=& \sum_{k=0}^\infty {}_kp_x v^{k} \\
&=& (\sum_{k=0}^0 + \sum_{k=1}^\infty) {}_kp_x v^{k} \\
&=& 1 + \sum_{k=1}^\infty {}_kp_x  v^{k} \\
& = & 1 + p_x \times v \times \ddot{a}_{x+1}
\end{eqnarray}

Hence

\begin{eqnarray}
\ddot{a}_x &=&  1 + p_x \times v \times \ddot{a}_{x+1}
\end{eqnarray}


**${}_nE_x$**

- Definition: $Z = v^{n} \times \mathbb{1}_{K \ge n}$
- Expected Value: ${}_nE_x = \mathbb{E}[Z] =  {}_np_x v^{n}$
- Expected Value expressed with Commutation Functions


Hence

\begin{eqnarray}
{}_nE_x &=&   p_x \times v \times {}_{n-1}E_{x+1}
\end{eqnarray}

In consequence we have

\begin{eqnarray}
\ddot{a}_x &= & \sum_{k=0}^\infty {}_kp_x v^x \\
     & = & \sum_{k=0}^\infty {}_kE_x
\end{eqnarray}

**$a_x$**

- Definition: $Y = \sum_{k=1}^\infty \mathbb{1}_{K\ge k} \times v^k$
- Expected Value: $a_x = \mathbb{E}[Y] = \sum_{k=1}^\infty \mathbb{E}[\mathbb{1}_{K\ge k} \times v^k] =
 \sum_{k=1}^\infty {}_kp_x \times v^k$


\begin{eqnarray}
Y & =&  \sum_{k=1}^\infty \mathbb{1}_{K\ge k} \times v^k \\
  & =&  \ddot{a}_K \\
  & = & v + \dots v^K \\
  & = & \frac{1-v^{K+1}}{1-v} -1 
\end{eqnarray}

- Expected Value: $a_x = \frac{1- \mathbb{E}[v^{K+1}]}{1-v} - 1 = \frac{1-A_x}{1-v} - 1 = \frac{v-A_x}{1-v}$

- Expected Value expressed with Commutation Functions
\begin{eqnarray}
a_x & = & \sum_{k=1} ^ \infty v^{k} {}_kp_x \\
    & = & \sum_{k=1} ^ \infty v^{k} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=1} ^ \infty \frac{v^{x+k}}{v^x} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=1} ^ \infty \frac{D_{x+k}}{D_x}\\
    & = & \frac{N_{x+1}}{D_x}
\end{eqnarray}
- Recusion between different years: Need to relate $A_x$ and $A_{x+1}$: You need ${}_kp_x = p_x \times {}_{k-1}p_{x+1}$
\begin{eqnarray}
a_x &=& \mathbb{E}[Y] \\ 
&=& \sum_{k=1}^\infty {}_kp_x v^{k} \\
&=& (\sum_{k=0}^0 + \sum_{k=1}^\infty) {}_kp_x v^{k} \\
&=& \sum_{k=1}^\infty {}_kp_x  v^{k} \\
& = & p_x \times v \times \ddot{a}_{x+1} \\
& = & p_x \times v \times (1+ a_{x+1})
\end{eqnarray}

Hence

\begin{eqnarray}
a_x &=&  p_x \times v \times (1+ a_{x+1})
\end{eqnarray}

**${}_m|\ddot{a}_x$**

- Definition: $Y = \sum_{k=m}^\infty \mathbb{1}_{K\ge k} \times v^k$
- Expected Value: ${}_m|\ddot{a}_x = \mathbb{E}[Y] = \sum_{k=m}^\infty \mathbb{E}[\mathbb{1}_{K\ge k} \times v^k] =
 \sum_{k=m}^\infty {}_kp_x \times v^k$


- Expected Value expressed with Commutation Functions
\begin{eqnarray}
{}_m|\ddot{a}_x & = & \sum_{k=m} ^ \infty v^{k} {}_kp_x \\
    & = & \sum_{k=m} ^ \infty v^{k} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=m} ^ \infty \frac{v^{x+k}}{v^x} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=m} ^ \infty \frac{D_{x+k}}{D_x}\\
    & = & \frac{N_{x+m}}{D_x} \\
    & = & \frac{N_{x+m}}{D_x} \frac{D_{x+m}}{D_{x+m}} \\
    & = & {}_mE_x \, \ddot{a}_{x+m}
\end{eqnarray}
- Recusion between different years: Need to relate $A_x$ and $A_{x+1}$: You need ${}_kp_x = p_x \times {}_{k-1}p_{x+1}$. For $m>1$
\begin{eqnarray}
{}_m|\ddot{a}_x &=& \mathbb{E}[Y] \\ 
&=& \sum_{k=m}^\infty {}_kp_x v^{k} \\
&=& (\sum_{k=0}^0 + \sum_{k=m}^\infty) {}_kp_x v^{k} \\
&=&  \sum_{k=1}^\infty {}_kp_x  v^{k} \\
& = &  p_x \times v \times {}_{m-1}|\ddot{a}_{x+1} 
\end{eqnarray}



**$\ddot{a}_{x:n}$**

- Definition: $Y = \sum_{k=0}^{n-1} \mathbb{1}_{K\ge k} \times v^k$
- Expected Value: $\ddot{a}_{x:n} = \mathbb{E}[Y] = \sum_{k=0}^{n-1} \mathbb{E}[\mathbb{1}_{K\ge k} \times v^k] =
 \sum_{k=0}^{n-1} {}_kp_x \times v^k$


- Expected Value expressed with Commutation Functions
\begin{eqnarray}
\ddot{a}_{x:n} & = & \sum_{k=0}^{n-1} v^{k} {}_kp_x \\
    & = & \sum_{k=0}^{n-1} v^{k} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=0}^{n-1} \frac{v^{x+k}}{v^x} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=0}^{n-1} \frac{D_{x+k}}{D_x}\\
    & = & \frac{N_x-N_{x+n}}{D_x}
\end{eqnarray}
- Recusion between different years: Need to relate $A_x$ and $A_{x+1}$: You need ${}_kp_x = p_x \times {}_{k-1}p_{x+1}$
\begin{eqnarray}
\ddot{a}_{x:n} &=& \mathbb{E}[Y] \\ 
&=& \sum_{k=0}^{n-1} {}_kp_x v^{k} \\
&=& (\sum_{k=0}^0 + \sum_{k=1}^{n-1}) {}_kp_x v^{k} \\
&=& 1 + \sum_{k=1}^{n-1} {}_kp_x  v^{k} \\
& = & 1 + p_x \times v \times \ddot{a}_{x+1:n-1}
\end{eqnarray}

Hence

\begin{eqnarray}
\ddot{a}_{x:n} &=&  1 + p_x \times v \times \ddot{a}_{x+1:n-1}
\end{eqnarray}

**General Annuity**

We are entitled to get an annuity $r_k \in \mathbb{R}$ if we are alive at this point in time. 
Given the current age is $x$ we need to look at   $Y = \sum_{k=0}^\infty r_{x+k} \mathbb{1}_{K\ge k} \times v^k$
- Expected Value: $\ddot{a}_x^{gen} = \mathbb{E}[Y] = \sum_{k=0}^\infty r_{x+k} \mathbb{E}[\mathbb{1}_{K\ge k} \times v^k] =
 \sum_{k=0}^\infty r_{x+k} {}_kp_x \times v^k$



- Expected Value expressed with Commutation Functions
\begin{eqnarray}
\ddot{a}_x^{gen} & = & \sum_{k=0} ^ \infty r_{x+k} v^{k} {}_kp_x \\
    & = & \sum_{k=0} ^ \infty r_{x+k} v^{k} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=0} ^ \infty  r_{x+k} \frac{v^{x+k}}{v^x} \frac{l_{x+k}}{l_x}  \\
     & = & \sum_{k=0} ^ \infty \frac{r_{x+k} \,D_{x+k}}{D_x}\\
\end{eqnarray}



- Recusion between different years: Need to relate $A_x$ and $A_{x+1}$: You need ${}_kp_x = p_x \times {}_{k-1}p_{x+1}$
\begin{eqnarray}
\ddot{a}_x^{gen} &=& \mathbb{E}[Y] \\ 
&=& \sum_{k=0}^\infty r_{x+k} \, {}_kp_x v^{k} \\
&=& (\sum_{k=0}^0 + \sum_{k=1}^\infty) r_{x+k} {}_kp_x v^{k} \\
&=& r_x + \sum_{k=1}^\infty {}_kp_x  v^{k} \\
& = & r_x + p_x \times v \times \ddot{a}_{x+1}^{gen}
\end{eqnarray}

Hence

\begin{eqnarray}
\ddot{a}_x^{gen} &=&  r_x + p_x \times v \times \ddot{a}_{x+1}^{gen}
\end{eqnarray}

with boundary condition $\ddot{a}_{\omega+1}^{gen} = 0$

**Example**

$r_{x+k} = (1+ \alpha)^k$

\begin{eqnarray}
\ddot{a}_x^{gen} & = & \sum_{k=0} ^ \infty (1+\alpha)^k v^{k} {}_kp_x \\
& = & \sum_{k=0} ^ \infty (\frac{1+\alpha}{1+i})^k {}_kp_x \\
& = & \frac{\tilde{N}_x}{\tilde{D}_x}
\end{eqnarray}

\begin{eqnarray}
\frac{1+\alpha}{1+i} & = & \frac{1}{1+\tilde{i}}
\end{eqnarray}


Using Taylor approximation $\tilde{i} \approx i - \alpha$
